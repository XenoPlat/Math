1) Пусть задана плоскость: Ax + By + Cz + D = 0

Напишите уравнение плоскости, параллельной данной и проходящей через начало координат.

Если x = y = z = 0, то D = 0. Тогда уравнение плоскости - Ax + By + Cz = 0

Пусть задана плоскость: A1x + B1y + C1z + D1 = 0 и прямая: (x - x1)/(x2 - x1) = (y - y1)/(y2 - y1) = (z - z1)/(z2 - z1). Как узнать, принадлежит прямая плоскости или нет?

Нужно сделать проверку: A1x1 + B1y1 + C1z1 + D1 = A1x2 + B1y2 + C1z2 + D1 = 0. Если тождество соблюдается, то прямая принадлежит плоскости. Данное утверждение основывается на условии принадлежности прямой к плоскости: хотябы 2 точки прямой должны приндлежать данной плоскости.
